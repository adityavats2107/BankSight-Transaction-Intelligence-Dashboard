### ***1️⃣ CUSTOMER & ACCOUNT ANALYSIS***
    Q1: How many customers exist per city, and what is their average account balance?

```python
import sqlite3
from tabulate import tabulate

# Connect to your SQLite database
conn = sqlite3.connect("bank_db")
cursor = conn.cursor()

query = """
SELECT 
    customers.city AS City,
    COUNT(customers.customer_id) AS Total_Customers,
    AVG(accounts.account_balance) AS Average_Account_Balance
FROM customers
JOIN accounts 
    ON customers.customer_id = accounts.customer_id
GROUP BY customers.city
ORDER BY Total_Customers DESC;
"""

cursor.execute(query)
rows = cursor.fetchall()

# Column names
columns = [description[0] for description in cursor.description]

# Print result in tabular format
print(tabulate(rows, headers=columns, tablefmt="grid"))
```

**Q2: Which account type (Savings, Current, Loan, etc.) holds the highest total balance?**

```python
query = """
SELECT 
    customers.account_type AS Account_Type,
    SUM(accounts.account_balance) AS Total_Balance
FROM customers
JOIN accounts 
    ON customers.customer_id = accounts.customer_id
GROUP BY customers.account_type
ORDER BY Total_Balance DESC
LIMIT 10;
"""

# Execute query
cursor.execute(query)
rows = cursor.fetchall()

# Extract column names
columns = [description[0] for description in cursor.description]

# Print table output
print(tabulate(rows, headers=columns, tablefmt="grid"))
```

**Q3: Who are the top 10 customers by total account balance across all account types?**

```python
query = """
SELECT 
    customers.customer_id,
    customers.name,
    SUM(accounts.account_balance) AS total_balance
FROM customers
JOIN accounts ON customers.customer_id = accounts.customer_id
GROUP BY customers.customer_id, customers.name
ORDER BY total_balance DESC
LIMIT 10;
"""

# 3. Execute and fetch
cursor.execute(query)
rows = cursor.fetchall()

# 4. Print results in table format
headers = [col[0] for col in cursor.description]
print(tabulate(rows, headers=headers, tablefmt="grid"))
```

**Q4: Which customers opened accounts in 2023 with a balance above ₹1,00,000?**

```python

query = """
SELECT 
    customers.customer_id,
    customers.name,
    customers.join_date,
    accounts.account_balance
FROM customers
JOIN accounts ON customers.customer_id = accounts.customer_id
WHERE strftime('%Y', customers.join_date) = '2023'
AND accounts.account_balance > 100000
ORDER BY accounts.account_balance DESC;
"""

cursor.execute(query)
rows = cursor.fetchall()
headers = [col[0] for col in cursor.description]
print(tabulate(rows, headers=headers, tablefmt="grid"))
```

# **2️⃣ TRANSACTION BEHAVIOR**

**Q5: What is the total transaction volume (sum of amounts) by transaction type?**

```python
query = """
SELECT 
    txn_type,
    SUM(amount) AS total_transaction_volume
FROM transactions
WHERE status = 'success'
GROUP BY txn_type
ORDER BY total_transaction_volume DESC;
"""

cursor.execute(query)
rows = cursor.fetchall()
headers = [col[0] for col in cursor.description]
print(tabulate(rows, headers=headers, tablefmt="grid"))
```

**Q6: Which accounts have more than 3 failed transactions in a single month?**

```python
query = """
SELECT 
    customers.name,
    transactions.customer_id,
    strftime('%Y-%m', transactions.txn_time) AS month,
    COUNT(*) AS failed_count
FROM transactions
JOIN customers ON transactions.customer_id = customers.customer_id
WHERE LOWER(TRIM(transactions.status)) = 'failed'
GROUP BY transactions.customer_id, month
HAVING failed_count > 3;
"""

cursor.execute(query)
rows = cursor.fetchall()
headers = [col[0] for col in cursor.description]

print(tabulate(rows, headers=headers, tablefmt="grid"))
```

**Q7: Which are the top 5 branches by total transaction volume in the last 6 months?**

```python
query = """
SELECT 
    branches.branch_name,
    SUM(transactions.amount) AS total_transaction_volume
FROM transactions
JOIN accounts ON transactions.txn_id = accounts.customer_id
JOIN customers ON accounts.customer_id = customers.customer_id
JOIN branches ON accounts.customer_id = branches.Branch_Id
WHERE transactions.txn_time >= DATE('now', '-6 months')
GROUP BY branches.Branch_Id
ORDER BY total_transaction_volume DESC
LIMIT 5;
"""

cursor.execute(query)
rows = cursor.fetchall()
headers = [col[0] for col in cursor.description]
print(tabulate(rows, headers=headers, tablefmt="grid"))
```

**Q8: Which accounts have 5 or more high-value transactions above ₹2,00,000?**

```python
query = """
SELECT 
    customer_id,
    COUNT(*) AS high_value_txns
FROM transactions
WHERE amount > 200000
AND LOWER(TRIM(status)) = 'success'
GROUP BY customer_id
HAVING high_value_txns >= 5;
"""

cursor.execute(query)
rows = cursor.fetchall()
headers = [col[0] for col in cursor.description]
print(tabulate(rows, headers=headers, tablefmt="grid"))
```

# 3️⃣ LOAN INSIGHTS

**Q9: What is the average loan amount and interest rate by loan type (Personal, Auto, Home, etc.)?**


```python
query = """
SELECT 
    loan_type,
    AVG(loan_amount) AS average_loan_amount,
    AVG(interest_rate) AS average_interest_rate
FROM loans
GROUP BY loan_type;
"""

cursor.execute(query)
rows = cursor.fetchall()
headers = [col[0] for col in cursor.description]
print(tabulate(rows, headers=headers, tablefmt="grid"))
```

**Q10: Which customers currently hold more than one active or approved loan?**


```python
query = """
SELECT 
    customer_id,
    COUNT(*) AS loan_count
FROM loans
WHERE LOWER(TRIM(loan_status)) IN ('active', 'approved')
GROUP BY customer_id
HAVING loan_count > 1;
"""

cursor.execute(query)
rows = cursor.fetchall()
headers = [col[0] for col in cursor.description]
print(tabulate(rows, headers=headers, tablefmt="grid"))
```

**Q11: Who are the top 5 customers with the highest outstanding (non-closed) loan amounts?**

```python
query = """
SELECT
    customer_id,
    SUM(loan_amount) AS total_outstanding_amount
FROM loans
WHERE LOWER(TRIM(loan_status)) != 'Closed'
GROUP BY customer_id
ORDER BY total_outstanding_amount DESC
LIMIT 5;
"""

cursor.execute(query)
rows = cursor.fetchall()
headers = [col[0] for col in cursor.description]
print(tabulate(rows, headers=headers, tablefmt="grid"))
```

# **4️⃣ BRANCH & PERFORMANCE**
**Q12: Which branch holds the highest total account balance?**

```python
query = """
SELECT 
    branches.branch_name,
    SUM(accounts.account_balance) AS total_balance
FROM accounts
JOIN customers 
    ON accounts.customer_id = customers.customer_id
JOIN branches
    ON customers.city = branches.city
GROUP BY branches.branch_name
ORDER BY total_balance DESC
LIMIT 1;
"""

cursor.execute(query)
rows = cursor.fetchall()
headers = [col[0] for col in cursor.description]

print(tabulate(rows, headers=headers, tablefmt="grid"))
```

**Q13: What is the branch performance summary showing total customers, total loans, and transaction volume?**

```python
query = """
SELECT 
    branches.branch_name,
    COUNT(DISTINCT customers.customer_id),
    (SELECT COUNT(*) FROM loans WHERE loans.Branch = branches.branch_name),
    (SELECT SUM(transactions.amount)
     FROM transactions
     JOIN customers ON transactions.customer_id = customers.customer_id
     WHERE customers.city = branches.city)
FROM branches
LEFT JOIN customers ON customers.city = branches.city
GROUP BY branches.branch_name
ORDER BY 
    (SELECT SUM(transactions.amount)
     FROM transactions
     JOIN customers ON transactions.customer_id = customers.customer_id
     WHERE customers.city = branches.city) DESC;
"""

cursor.execute(query)
rows = cursor.fetchall()
headers = ["branch_name", "total_customers", "total_loans", "total_transaction_volume"]

print(tabulate(rows, headers=headers, tablefmt="grid"))
```

# **5️⃣ SUPPORT TICKETS & CUSTOMER EXPERIENCE**

**Q14: Which issue categories have the longest average resolution time?**

```python
query = """
SELECT 
    Issue_Category,
    COUNT(*) AS total_tickets,
    ROUND(AVG(JULIANDAY(Date_Closed) - JULIANDAY(Date_Opened)), 2) AS avg_resolution_days
FROM support_tickets
WHERE Status IN ('Closed', 'Resolved')
  AND Date_Closed IS NOT NULL 
  AND Date_Opened IS NOT NULL
GROUP BY Issue_Category
ORDER BY avg_resolution_days DESC;
"""

cursor.execute(query)
rows = cursor.fetchall()
headers = [col[0] for col in cursor.description]
print(tabulate(rows, headers=headers, tablefmt="grid"))
```

**Q15: Which support agents have resolved the most critical tickets with high customer ratings (≥4)?**


```python
query = """
SELECT 
    Support_Agent,
    COUNT(*) AS critical_tickets_resolved,
    ROUND(AVG(Customer_Rating), 2) AS avg_customer_rating
FROM support_tickets
WHERE Priority = 'Critical'
  AND Status IN ('Resolved', 'Closed')
  AND Customer_Rating >= 4
GROUP BY Support_Agent
ORDER BY critical_tickets_resolved DESC, avg_customer_rating DESC;
"""
cursor.execute(query)
rows = cursor.fetchall()
headers = [col[0] for col in cursor.description]
print(tabulate(rows, headers=headers, tablefmt="grid"))
```

# **Step 5: Remove all duplicacy from all tables in bank_db**

```python
tables = cursor.execute(
    "SELECT name FROM sqlite_master WHERE type='table' AND name NOT LIKE 'sqlite_%';"
).fetchall()

for (table,) in tables:
    print(f"Cleaning duplicates in: {table}")

    cursor.executescript(f"""
        CREATE TABLE temp_clean AS 
        SELECT DISTINCT * FROM {table};

        DROP TABLE {table};

        ALTER TABLE temp_clean RENAME TO {table};
    """)

conn.commit()
conn.close()

print("✔ All tables cleaned. No duplicates remain.")
```
